## Menentukan Pertanyaan Bisnis

**SMART Question** adalah sebuah framework untuk merumuskan pertanyaan secara terstruktur agar memeperoleh informasi yang mendalam.

* **Spesific (Spesifik)**
  * Pertanyaan harus jelas, fokus pada sebuah topik tertentu, dan tidak bermakna ganda. Hindari pertanyaan yang terlalu luas.
    * Salah: Bagaimana cara meningkatkan penjualan?
    * Benar: Faktor apa saja yang memengaruhi penurunan penjualan produk kategori elektronik di wilayah Jakarta selama kuartal terakhir?
* **Measurable (Terukur)**
  * Pertanyaan harus bisa dijawab dengan angka atau matrix yang konkret, harus tahu apa yang akan dihitung.
    * Salah: Apakah pelanggan senang dengan layanan kita?
    * Benar: Berapa skor rata-rata Customer Satisfaction untuk layanan purna jual bulan ini dibandingkan bulan lalu?
* **Action-Oriented (Berorientasi Aksi)**
  * Hasil dari pertanyaan harus bisa memberikan arahan untuk melakukan tindakan nyata. Jika pertanyaan terjawab, stakeholder harus tahu apa langkah selanjutnya.
    * Salah: Mengapa orang suka berbelanja?
    * Benar: Fitur apa pada aplikasi yang paling sering digunakan sebelum pengguna memutuskan untuk melakukan checkout?
* **Relevant (Relevan)**
  * Hasil dari pertanyaan harus sejalan dengan tujuan utama bisnis atau masalah yang sedang dihadapi.
    * Salah: Menanyakan tentang stok gudang saat masalah utamanya adalah efektivitas kampanye media sosial.
    * Benar: Apakah kampanye iklan di Instagram memberikan Return on Ad Spend (ROAS) yang lebih tinggi dibandingkan iklan di TikTok?
* **Time-bound (Terikat Waktu)**
  * Pertanyaan harus ada batasan waktu yang jelas agar analisis memiliki konteks yang tepat.
    * Salah: Berapa banyak pengguna baru kita?
    * Benar: Berapa tingkat pertumbuhan pengguna baru secara bulanan (Month-over-Month) sepanjang tahun 2025?

**Contoh pertanyaan bisnis yang memenuhi seluruh elemen SMART**

***"Faktor apa saja yang memengaruhi penurunan conversion rate pada pengguna aplikasi Android di wilayah Jabodetabek sebesar 5% selama periode Flash Sale Maret 2026?"***

Keterangan:

- **Specific**: Fokus pada "penurunan conversion rate" untuk "aplikasi Android" di "Jabodetabek". Bukan sekadar penjualan turun.
- **Measurable**: Ada angka konkret yang ingin dianalisis, yaitu penurunan sebesar "5%".
- **Action-Oriented**: Dengan mengetahui faktor penyebabnya misalnya bug pada versi Android tertentu atau kendala logistik di Jabodetabek, tim bisa langsung melakukan perbaikan teknis atau operasional.
- **Relevant**: Penurunan konversi saat Flash Sale adalah masalah kritis bagi bisnis retail/e-commerce.
- **Time-bound**: Dibatasi pada periode spesifik "Flash Sale Maret 2026".

- **Pertanyaan 1:** **Bagaimana pemetaan 5 kategori produk dengan volume penjualan tertinggi dan terendah selama periode tahun 2018?**
   - **Specific:** Fokus memetakan 5 produk terbaik dan terburuk untuk prioritas pengadaan dan efisiensi kapasitas penyimpanan di gudang.
   - **Measurable:** Diukur menggunakan total transaksi sukses pada tiap kategori produk.
   - **Action-Oriented:** Menjadi dasar keputusan penambahan stok atau cuci gudang.
   - **Relevant:** Penting untuk mengoptimalkan biaya penyimpanan gudang.
   - **Time-bound:** Dibatasi pada transaksi sepanjang tahun 2018.

- **Pertanyaan 2:** **Bagaimana tren performa jumlah pesanan bulanan dan total pendapatan (*revenue*) perusahaan selama periode januari-agustus 2018?**
   - **Specific & Time-bound:** Fokus pada tren pesanan dan pendapatan bulanan khusus pada periode januari-agustus tahun 2018.
   - **Measurable:** Dapat diukur melalui jumlah pesanan tervalidasi dan total nilai transaksi (akumulasi harga produk dan biaya pengiriman) di setiap bulannya.
   - **Action-oriented**: Digunakan untuk mengevaluasi efektivitas promosi bulanan dan merancang alokasi anggaran kampanye pemasaran di tahun berikutnya
   - **Relevant**: Krusial untuk memantau apakah bisnis sedang bertumbuh atau mengalami penurunan secara finansial.
   **Time-bound**: Dibatasi secara spesifik hanya pada periode historis dari Januari hingga Agustus 2018.

### Cleaning Data

#### Fixing problem

```python
# 1. Customers Dataset
# Data sudah 100% bersih dari missing value dan duplikat,Tipe data sudah sesuai (object/string), dan tidak ada outlier karena data kategorik. jadi tidak ada aksi pembersihan pada dataset ini.
print("--- Cek Dataset Customers ---")
print(f"Dimensi data : {customers_df.shape}")
print(f"Missing value: {customers_df.isna().sum().sum()}")
```

```python
# 2. Orders Dataset
# Menyaring hanya pesanan yang sukses terkirim (delivered)
orders_df = orders_df[orders_df["order_status"] == "delivered"]

# Mengubah tipe data kolom waktu dari object ke datetime
orders_df["order_purchase_timestamp"] = pd.to_datetime(orders_df["order_purchase_timestamp"])

# Membatasi data agar hanya mencakup transaksi di tahun 2018
orders_df = orders_df[orders_df["order_purchase_timestamp"].dt.year == 2018]

# Menghapus kolom operasional yang tidak diperlukan untuk analisis pendapatan
orders_df.drop(columns=["order_status", "order_approved_at", "order_delivered_carrier_date",
                        "order_delivered_customer_date", "order_estimated_delivery_date"], inplace=True)
```

```python
# Cek Hasil Cleaning Orders
print("Dimensi data akhir:", orders_df.shape)
print("Tipe data kolom timestamp:", orders_df['order_purchase_timestamp'].dtype)
print("Daftar tahun unik:", orders_df['order_purchase_timestamp'].dt.year.unique())
print("Daftar kolom tersedia:", orders_df.columns.tolist())
```

```python
# 3. Items Dataset
# Menghapus kolom seller_id dan shipping_limit_date yang tidak relevan
items_df.drop(columns=["seller_id", "shipping_limit_date"], inplace=True)

# Outlier pada harga dan ongkos kirim tetap dipertahankan karena merupakan transaksi yang sah
```

```python
# Cek Hasil Cleaning Items
print("Daftar kolom tersedia:", items_df.columns.tolist())
print("Nilai harga tertinggi:", items_df['price'].max())
```

```python
# 4. Products Dataset dan Translation dataset
# Hapus kolom atribut fisik yang banyak outlier ekstrem dan tidak relevan dalam menjawab pertanyaan bisnis
products_df.drop(columns=["product_name_lenght", "product_description_lenght", "product_photos_qty",
                          "product_weight_g", "product_length_cm", "product_height_cm", "product_width_cm"], inplace=True)

# Mengubah nama kategori produk asli dari dataset produk dan nama kategori produk bahasa inggris dari dataset translasi menjadi huruf kecil semua dan mengganti underscore dengan spasi
products_df["product_category_name"] = products_df["product_category_name"].str.lower().str.replace("_", " ")
translation_df["product_category_name"] = translation_df["product_category_name"].str.lower().str.replace("_", " ")

# Isi data nama kategori produk yang kosong dengan 'unknown'
products_df["product_category_name"] = products_df["product_category_name"].fillna("unknown")

# Menggabungkan data nama kategori produk dari dataset translation_df ke dalam dataset products_df
products_df = pd.merge(products_df, translation_df, on="product_category_name", how="left")

# Jika terjemahan Inggris tidak ditemukan, gunakan nama kategori asli (dalam bahasa portugis) agar tidak kosong
products_df["product_category_name_english"] = products_df["product_category_name_english"].fillna(products_df["product_category_name"])

# 5. Drop kolom lama dan Rename kolom baru menjadi nama yang lebih netral
products_df.drop(columns=["product_category_name"], inplace=True)
products_df.rename(columns={"product_category_name_english": "product_category"}, inplace=True)

# Verifikasi hasil akhir satu baris
print("Cleaning Produk Selesai. Kolom saat ini:", products_df.columns.tolist())
```

## Analisis Lanjutan (Opsional)

```python
# 1. Membuat kolom baru `total_revenue` dengan menjumlahkan `price` dan `freight_value` untuk mendapatkan total nilai transaksi.
main_df['total_revenue'] = main_df['price'] + main_df['freight_value']
latest_date = main_df['order_purchase_timestamp'].max() + pd.Timedelta(days=1)
rfm_df = main_df.groupby('customer_unique_id').agg({
    'order_purchase_timestamp': lambda x: (latest_date - x.max()).days,
    'order_id': 'count',
    'total_revenue': 'sum'
}).reset_index()

# 2. Menghitung Recency, Frequency, dan Monetary
# Latest date diset sehari setelah transaksi terakhir di dataset
latest_date = main_df['order_purchase_timestamp'].max() + pd.Timedelta(days=1)

rfm_df = main_df.groupby('customer_unique_id').agg({
    'order_purchase_timestamp': lambda x: (latest_date - x.max()).days, # Recency
    'order_id': 'count',                                               # Frequency
    'total_revenue': 'sum'                                             # Monetary
}).reset_index()

rfm_df.columns = ['customer_id', 'recency', 'frequency', 'monetary']

# 3. Scoring (1-5) menggunakan Quantiles
rfm_df['r_score'] = pd.qcut(rfm_df['recency'], 5, labels=[5, 4, 3, 2, 1])
rfm_df['f_score'] = pd.qcut(rfm_df['frequency'].rank(method='first'), 5, labels=[1, 2, 3, 4, 5])
rfm_df['m_score'] = pd.qcut(rfm_df['monetary'], 5, labels=[1, 2, 3, 4, 5])

# 4. Segmentasi Berdasarkan Total Skor
rfm_df['total_score'] = rfm_df[['r_score', 'f_score', 'm_score']].astype(int).sum(axis=1)

def assign_segment(score):
    if score >= 12: return 'Top Customers'
    if score >= 9:  return 'Loyal Customers'
    if score >= 6:  # Gabungan segmen berisiko
        return 'At Risk'
    return 'Lost Customers'

rfm_df['segment'] = rfm_df['total_score'].apply(assign_segment)

# 5. Visualisasi Donut Chart
segment_counts = rfm_df['segment'].value_counts()
total_customers = segment_counts.sum()

# Menyiapkan label dengan format: "Nama Segmen \n Persentase"
labels = [f"{index}\n({(value/total_customers)*100:.1f}%)" for index, value in zip(segment_counts.index, segment_counts.values)]
colors = ["#2ecc71", "#3498db", "#f1c40f", "#e74c3c"]

plt.figure(figsize=(12, 10))

# Plot Pie Chart
plt.pie(
    segment_counts,
    labels=labels,
    startangle=140,
    colors=colors,
    labeldistance=1.15,
    explode=[0.05] * len(segment_counts),
    textprops={'fontsize': 12, 'fontweight': 'bold', 'ha': 'center'}
)

centre_circle = plt.Circle((0,0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("Komposisi Profil Pelanggan Olist 2018 (RFM Analysis)", fontsize=20, fontweight='bold', pad=40)
plt.tight_layout()
plt.show()
```

Secara keseluruhan, kondisi pelanggan sebenarnya sangat sehat dimana gabungan antara Loyal Customers (39,4%) dan Top Customers (17,4%) menunjukkan bahwa lebih dari 56% pelanggan adalah pembeli aktif yang percaya dengan platform perusahaan. Mereka adalah tulang punggung yang menjaga revenue tetap stabil di angka jutaan BRL setiap bulannya.

Namun meskipun demikian, kelompok At Risk (34,4%) adalah poin yang paling butuh perhatian. Ini adalah pelanggan yang dulunya rutin belanja tapi sekarang mulai jarang terlihat. Angka sepertiga dari total pelanggan ini adalah sebuah peringatan yang apabila dibiarkan, maka mereka akan segera pindah ke segmen Lost, yang artinya perusahaan bakal kehilangan potensi pendapatan dalam jumlah besar.

Berita bagusnya, jumlah Lost Customers kita tergolong kecil, hanya 8,8%. Ini artinya perusahaan punya daya tarik yang cukup kuat untuk membuat orang tidak langsung kapok setelah belanja pertama kali.
